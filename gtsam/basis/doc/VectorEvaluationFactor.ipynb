## Overview

`VectorEvaluationFactor` is a unary factor for vector-valued measurements evaluated from a basis parameter matrix. It enforces $\mathbf{f}(x)$ to match a measurement vector at a given point.

## Key Functionality / API

- `VectorEvaluationFactor(key, z, model, M, N, x)` constrains the full vector.
- Optional interval parameters `a, b` scale the basis domain.


## Usage Example

```cpp
gtsam::Vector z = (gtsam::Vector(3) << 1.0, 0.0, -0.5).finished();
auto model = gtsam::noiseModel::Isotropic::Sigma(3, 0.1);
size_t M = 3, N = 6;
double x = 0.0;
gtsam::VectorEvaluationFactor<gtsam::Chebyshev2> factor(key, z, model, M, N, x);
```
