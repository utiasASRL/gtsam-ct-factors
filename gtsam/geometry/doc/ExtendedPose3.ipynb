# ExtendedPose3

`ExtendedPose3<K>` is the generic Lie group $SE_K(3)$: the semi-direct product of `SO(3)` with `K` copies of $\mathbb{R}^3$.

It is a matrix Lie group with homogeneous representation size $(3+K)\times(3+K)$ and manifold dimension $3 + 3K$.

## State and Tangent Layout

An element is parameterized as:

$$
X = (R, x_1, \ldots, x_K), \qquad R \in SO(3),\; x_i \in \mathbb{R}^3
$$

with tangent vector ordering:

$$
\xi = [\omega, \rho_1, \ldots, \rho_K] \in \mathbb{R}^{3+3K}.
$$

## Matrix Representation

$$
X = \begin{bmatrix}
R & x_1 & \cdots & x_K \\
0 & 1 & & 0 \\
\vdots & & \ddots & \\
0 & 0 & & 1
\end{bmatrix}
$$

The group operation is:

$$
(R,x_i)\,(S,y_i) = (RS,\; x_i + R y_i).
$$

## C++ Usage Notes

- Fixed-size: `ExtendedPose3<3>` gives a 12D manifold.
- Dynamic-size: `ExtendedPose3<Eigen::Dynamic>` takes runtime `k` in constructor.
- Default dynamic construction (`k=0`) is supported for serialization and placeholder use.
- The API provides the standard `Manifold`, `LieGroup`, and `MatrixLieGroup` surface: `retract/localCoordinates`, `Expmap/Logmap`, `AdjointMap`, `Hat/Vee`, and `matrix/vec`.
